```python
#Question 4, part c
import numpy as np  
import matplotlib.pyplot as plt #needed to plot

mean = 171 #avergae height
variance = 7.1**2  #variance in height
a = 162
b = 190

def normal_density(mean, variance, x):
    function = (1/(np.sqrt(2*np.pi*(variance))))*(np.exp(-((x-mean)**2)/(2*(variance)))) #put your code here
    return function # the value of your function.

def integration(mean, variance, a, b, n=1000000):
    w = (b-a)/n
    midpoints = np.linspace(a + w/2, b - w/2, n)
    y = normal_density(mean, variance, midpoints)
    integral = np.sum(y * w)
    return(integral)

def average_dosage(mean, variance, a, b, n=1000000):
    dosage = lambda x: 2.38 * x**2  # D(x) = 2.38 * X^2
    w = (b-a)/n
    midpoints = np.linspace(a + w/2, b - w/2, n)
    y = normal_density(mean, variance, midpoints)
    dosage = dosage(midpoints) 
    expected_dosage = np.sum(dosage * y * w)
    return(expected_dosage)

expected_dosage = average_dosage(mean, variance, a, b)

print("Estimated average dosage: ", expected_dosage)
    
```
